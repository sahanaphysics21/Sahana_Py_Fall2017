## Blocking Functions: Show and Input by the user at the keyboard


```python
# Alternative program to write sinx

import numpy as np
from pylab import plot, show, ylim, xlim, xlabel, ylabel

xpoints =[]
y1points =[]                                     # Defining empty vectors
y2points =[]
y3points =[]

for x in np.linspace(0,10,100):
    xpoints.append(x)                           # Introduces values in the range of the for loop 
    y1points.append(np.sin(x))
    y2points.append(np.cos(x))
    y3points.append(np.sin(x**2))
    
plot(xpoints,y1points, "g--")                     # Dotted and color
plot(xpoints,y2points, "co")                      # Circular 'o', color cyan 'c'
plot(xpoints,y3points, "y")                       # Solid graph

ylim(-2.1,1.1)                                    # Introduces space
xlabel("Time")                                   # Axes name
ylabel("Y")
show()
```

### Exercise 3.2: Deltoid Curve

```python
import numpy as np
from pylab import plot,show

# Writing equation for Deltoid Curve for the Given Data
theta = np.linspace(0,2*pi,6280)
x = 2*(np.cos(theta))+np.cos(2*theta)
y = 2*(np.sin(theta))-np.sin(2*theta)

plot(x,y,"r")
show()
```

### Exercise 3.2: Fey Function 

```python
import numpy as np
from pylab import plot,show

x = []
y = []
for theta in linspace(0,24*pi,31400):
    a = np.cos(theta)
    b = (np.sin(theta/12))**5
    r = np.exp(a)-2*np.cos(4*theta)+b
    x.append(r*cos(theta))
    y.append(r*sin(theta))

plt.plot(x,y,"b--")  
plt.show()
```
